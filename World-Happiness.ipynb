```python
grouped = pd.concat([first["Country"], first["Region"],
                    first["Happiness Score"], first["Happiness Rank"], 
                    second["Happiness Score"], second["Happiness Rank"], 
                    third["Happiness Score"], third["Happiness Rank"], 
                    fourth["Happiness Score"], fourth["Happiness Rank"], 
                    fifth["Happiness Score"], fifth["Happiness Rank"]],
                    axis=1, 
                    keys=["Country", "Region", 
                          "Happiness Score 2015", "Happiness Rank 2015",
                          "Happiness Score 2016", "Happiness Rank 2016",
                          "Happiness Score 2017", "Happiness Rank 2017", 
                          "Happiness Score 2018", "Happiness Rank 2018", 
                          "Happiness Score 2019", "Happiness Rank 2019"]
                   )


```

```python
grouped.head()
```

```python
grouped[pd.isnull(grouped).any(axis=1)]
grouped.dropna()
```

```python
grouped["Rank Change"] = grouped["Happiness Rank 2019"] - grouped["Happiness Rank 2015"]
grouped["Score Change"] = grouped["Happiness Score 2019"] - grouped["Happiness Score 2015"]
```

```python
sorted_by_rankChange = grouped.sort_values(by=["Rank Change"], ascending=False)
sorted_by_rankChange.head()
```

```python
sorted_by_scoreChange = grouped.sort_values(by=["Score Change"], ascending=True)
sorted_by_scoreChange.head()
```

```python
grouped_region = grouped.groupby(["Region"])[["Happiness Score 2015", "Happiness Score 2016", 
                                    "Happiness Score 2017", "Happiness Score 2018", "Happiness Score 2019"]]
```

```python
grouped_region.groups.keys()
```

```python
from matplotlib.ticker import FormatStrFormatter

fig, axs = plt.subplots(3,2, figsize=(15,15))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.6)

axs[0,0].bar(grouped_region.groups.keys(), grouped_region.mean()["Happiness Score 2015"])
axs[0,0].tick_params(axis='x', rotation=70)
axs[0,0].set_xlabel("Score")
axs[0,0].set_ylabel("Region")

axs[0,1].bar(grouped_region.groups.keys(), grouped_region.mean()["Happiness Score 2016"])
axs[0,1].tick_params(axis='x', rotation=70)
axs[0,1].set_xlabel("Region")
axs[0,1].set_ylabel("Score")

axs[1,0].bar(grouped_region.groups.keys(), grouped_region.mean()["Happiness Score 2017"])
axs[1,0].tick_params(axis='x', rotation=70)
axs[1,0].set_xlabel("Region")
axs[1,0].set_ylabel("Score")

axs[1,1].bar(grouped_region.groups.keys(), grouped_region.mean()["Happiness Score 2018"])
axs[1,1].tick_params(axis='x', rotation=70)
axs[1,1].set_xlabel("Region")
axs[1,1].set_ylabel("Score")

axs[2,0].bar(grouped_region.groups.keys(), grouped_region.mean()["Happiness Score 2019"])
axs[2,0].tick_params(axis='x', rotation=70)
axs[2,0].set_xlabel("Score")
axs[2,0].set_ylabel("Region")

plt.show()
```

```python
dataframe_mean = pd.concat([first, second, third, fourth, fifth]).groupby("Country", as_index=False).mean()
```

```python
dataframe_mean.sort_values(by="Happiness Score", ascending=False, inplace=True)
dataframe_mean.index = np.arange(1, len(dataframe_mean)+1)
```

```python
dataframe_mean.head(10)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(18, 12))

sns.barplot(ax=axs[0, 0], data=dataframe_mean.head(10), x="Country", y="Happiness Score")
sns.barplot(ax=axs[0, 1], data=dataframe_mean.sort_values(by="Economy (GDP per Capita)", ascending=False).head(10), x="Country", y="Economy (GDP per Capita)")
sns.barplot(ax=axs[1, 0], data=dataframe_mean.sort_values(by="Family", ascending=False).head(10), x="Country", y="Family")
sns.barplot(ax=axs[1, 1], data=dataframe_mean.sort_values(by="Health (Life Expectancy)", ascending=False).head(10), x="Country", y="Health (Life Expectancy)")
sns.barplot(ax=axs[2, 0], data=dataframe_mean.sort_values(by="Freedom", ascending=False).head(10), x="Country", y="Freedom")
sns.barplot(ax=axs[2, 1], data=dataframe_mean.sort_values(by="Trust (Government Corruption)", ascending=False).head(10), x="Country", y="Trust (Government Corruption)")


axs[0,0].tick_params(axis='x', rotation=15)
axs[0,1].tick_params(axis='x', rotation=15)
axs[1,0].tick_params(axis='x', rotation=15)
axs[1,1].tick_params(axis='x', rotation=15)
axs[2,0].tick_params(axis='x', rotation=15)
axs[2,1].tick_params(axis='x', rotation=15)
```
